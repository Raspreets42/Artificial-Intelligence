```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
}, index=['Row1', 'Row2', 'Row3', 'Row4'])

print(df)
```

```python
print(f"\nShape: {df.shape}")
print(f"Axis 0 (rows): {df.shape[0]} rows")
print(f"Axis 1 (columns): {df.shape[1]} columns")
```

```python
print(f"Mean of each column (axis=0): \n{df.mean(axis=0)}")
```

```python
print(f"Mean of each rows (axis=1): \n{df.mean(axis=1)}")
```

```python
print("\nSum of each column (axis=0):")
print(df.sum(axis=0))

print("\nSum of each row (axis=1):")
print(df.sum(axis=1))
```

```python
df_rows_dropped = df.drop('Row2', axis=0)
print("After dropping Row2 (axis=0):")
print(df_rows_dropped)
```
